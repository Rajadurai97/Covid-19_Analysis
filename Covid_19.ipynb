```
datafr = pd.DataFrame(data,columns=['Date','Total_Cases'])
datafr
```

```
#Line graph of data wise cases
fig, ax = plt.subplots()
fig.set_figwidth(20)
fig.set_figheight(8)
ax.set_xlabel('Date',fontsize=20)
plt.xticks(rotation=50)
ax.set_ylabel('Total_Cases')
ax.plot(datafr['Date'],datafr['Total_Cases'])
plt.title('Covid-19 Date wise cases',fontsize=25)
plt.ylabel('Number of cases',fontsize=20)
plt.show()
```

```
#Bar graph of data wise cases
fig, ax = plt.subplots()
fig.set_figwidth(20)
fig.set_figheight(8)
Date = datafr['Date']
Cases = datafr['Total_Cases']
sns.barplot(x=Date,y=Cases,palette='tab10')
plt.xticks(rotation=50)
plt.title('Covid-19 Date wise cases',fontsize=25)
plt.ylabel('Number of cases',fontsize=20)
plt.show()
```
